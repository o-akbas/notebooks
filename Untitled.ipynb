```python
def ricker(f, length=0.512, dt=0.001):
    t = np.arange(-length/2, (length-dt)/2, dt)
    y = (1.-2.*(np.pi**2)*(f**2)*(t**2))*np.exp(-(np.pi**2)*(f**2)*(t**2))
    return t, y
```
